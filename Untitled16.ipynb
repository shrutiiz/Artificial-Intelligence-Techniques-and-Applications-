```python
from ultralytics import YOLO

# Load a YOLOv8 model (YOLOv8n is the smallest; you can also use YOLOv8s, YOLOv8m, etc.)
model = YOLO('yolov8n.pt')  # Pretrained YOLOv8 weights

# Train the model
model.train(data='/content/drive/MyDrive/Colab Notebooks/Crack_detection.v8i.yolov8/data.yaml',
            epochs=3,
            batch=16,
            imgsz=203)
```
